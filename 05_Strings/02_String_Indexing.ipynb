## Description

String indexing is used to access individual characters from a string.

Each character has its own position called an index.

## What is String Indexing?

Indexing means accessing a specific character of a string using its position.

Python provides two types of indexing:

- Positive Indexing
- Negative Indexing

## Positive Indexing

Positive indexing starts from **0** and moves from left to right.

Example:

```
P  Y  T  H  O  N
0  1  2  3  4  5
```

````python
language = "PYTHON"

print(language[0])
print(language[1])
print(language[2])
print(language[3])
print(language[4])
print(language[5])
````

## Negative Indexing

Negative indexing starts from **-1** and moves from right to left.

Example:

```
 P  Y  T  H  O  N
-6 -5 -4 -3 -2 -1
```

## Negative Indexing

Negative indexing starts from **-1** and moves from right to left.

Example:

```
 P  Y  T  H  O  N
-6 -5 -4 -3 -2 -1
```

````python
language = "PYTHON"

print(language[-1])
print(language[-2])
print(language[-3])
print(language[-4])
print(language[-5])
print(language[-6])
````

## Accessing First and Last Character

````python
word = "Programming"

print("First Character :", word[0])
print("Last Character :", word[-1])
````

## Accessing Middle Characters

````python
city = "Pune"

print(city[1])
print(city[2])
````

## Using Variables for Indexing

````python
name = "Shilpa"

index = 3

print(name[index])
````

## Index Out of Range

````python
text = "Python"

# This will generate an error
print(text[10])
````

````python
text = "Python"
print(text[1])
````

## Real-Life Examples

Indexing is useful for:

- Getting the first letter of a name
- Getting the last digit of a phone number
- Extracting initials
- Validating file extensions

````python
email = "student@gmail.com"

print("First Character :", email[0])
print("Last Character :", email[-1])
````

## Data Analytics Connection

Indexing helps while:

## Common Mistakes

- Forgetting indexing starts at 0.
- Using an index larger than the string length.
- Confusing positive and negative indexing.

## Key Points

- Positive indexing starts from 0.
- Negative indexing starts from -1.
- Every character has an index.
- Invalid indexes generate an IndexError.

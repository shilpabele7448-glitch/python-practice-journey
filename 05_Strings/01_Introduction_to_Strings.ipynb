## What is a String?

A string is a sequence of characters enclosed in:

- Single quotes (' ')
- Double quotes (" ")
- Triple quotes (''' ''' or """ """)

Examples:

- "Python"
- "Hello World"
- "Data Analytics"
- "Shilpa"

Strings are **immutable**, which means they cannot be changed after they are created.

```python
name = "Shilpa"

print(name)
```

```python
language = 'Python'

print(language)
```

```python
message = """Welcome to Python.

Python is easy to learn.
Python is powerful.
"""

print(message)
```

```python
city = "Pune"

print(type(city))
```

```python
word = "Python"

print("Length =", len(word))
```

```python
first_name = "Shilpa"
last_name = "Bele"

full_name = first_name + " " + last_name

print(full_name)
```

```python
word = "Hi "

print(word * 5)
```

```python
course = "Python Programming"

print(course)
print(type(course))
print(len(course))
```

```python
quote = "Learning Python is fun!"

print(quote)
```

```python
email = "shilpa@example.com"

print(email)
```

```python
password = "Python@123"

print(password)
```

## Rules for Strings

- Strings must be enclosed in quotes.
- Strings can contain letters, numbers, spaces, and special characters.
- Strings are immutable.
- Strings support many built-in methods.

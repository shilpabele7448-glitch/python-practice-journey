## Why is this Topic Important?

Many programs need to make decisions.

For example:
- Is a person eligible to vote?
- Is the password correct?
- Did the student pass?

The if statement helps Python make these decisions.

## What is an If Statement?

An if statement checks whether a condition is True.

If the condition is True, Python executes the code inside the if block.

If the condition is False, Python skips that block.

## Syntax

if condition:
    statement

```python
age = 20

if age >= 18:
    print("You are eligible to vote.")
```

```python
marks = 75

if marks >= 35:
    print("You Passed!")
```

```python
temperature = 40

if temperature > 35:
    print("It's a hot day.")
```

```python
number = 8

if number % 2 == 0:
    print("Even Number")
```

```python
password = "python123"

user_password = input("Enter Password: ")

if user_password == password:
    print("Login Successful")
```

```python
salary = int(input("Enter Salary: "))

if salary > 50000:
    print("High Salary")
```

## Flow of an If Statement

Condition → True?

Yes → Execute Code

No → Skip Code

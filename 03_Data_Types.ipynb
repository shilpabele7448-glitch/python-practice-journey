## What are Data Types?

A data type tells Python what kind of value is stored in a variable. Different data types are used to store different kinds of information.

For example:
- Age is a whole number.
- Salary may contain decimal values.
- Name is text.
- Pass/Fail is either True or False.

```python
age = 22

print(age)
print(type(age))
```

```python
salary = 45000.75

print(salary)
print(type(salary))
```

```python
name = "Shilpa"

print(name)
print(type(name))
```

```python
is_student = True

print(is_student)
print(type(is_student))
```

## Four Basic Data Types

### int
Stores whole numbers.

Example:
10, 20, -5

---

### float
Stores decimal numbers.

Example:
99.99, 12.5

---

### str
Stores text inside quotes.

Example:
"Python"

---

### bool
Stores only two values:

True

False

```python
number = 100
price = 99.99
language = "Python"
is_pass = False

print(type(number))
print(type(price))
print(type(language))
print(type(is_pass))
```

## Why are Data Types Important?

Data types help Python understand what kind of operation can be performed on a value.

For example:
- Numbers can be added.
- Strings can be joined.
- Boolean values are used in conditions.

```python
a = 10
b = 20

print(a + b)
```

```python
first = "Data"
second = "Analytics"

print(first + " " + second)
```

```python
numbers = [10, 20, 30]
print(type(numbers))
```

```python
marks = (75, 80, 90)
print(type(marks))
```

```python
colors = {"Red", "Blue", "Green"}
print(type(colors))
```

```python
student = {"name":"Rahul","age":21}
print(type(student))
```

```python
value = None
print(type(value))
```

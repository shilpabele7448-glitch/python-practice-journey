## What is a Variable?

A variable is a name used to store data in a program. Instead of writing the same value again and again, we store it in a variable and use its name whenever needed.

### Real-Life Example

Think of a variable as a labeled container.

If a container is labeled **Age**, it can store the value **22**.

If it is labeled **Name**, it can store **Shilpa**.

```python
name = "Shilpa"

print(name)
```

```python
age = 22

print(age)
```

```python
city = "Pune"

print(city)
```

```python
price = 499.99

print(price)
```

```python
is_student = True

print(is_student)
```

## Variable Naming Rules

✅ Variable names can contain letters, numbers, and underscores.

✅ Variable names cannot start with a number.

✅ Variable names are case-sensitive.

✅ Do not use Python keywords as variable names.

```python
first_name = "Shilpa"

print(first_name)
```

```python
age = 22
Age = 30

print(age)
print(Age)
```

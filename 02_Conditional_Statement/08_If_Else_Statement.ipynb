## Why is this Topic Important?

Sometimes there are only two possible outcomes.

For example:
- Pass or Fail
- Even or Odd
- Login Successful or Login Failed

The if...else statement helps Python choose between two options.

## What is an If...Else Statement?

The if statement checks a condition.

If the condition is True, the if block executes.

If the condition is False, the else block executes.

## Syntax

if condition:
    statement
else:
    statement

```python
marks = 45

if marks >= 35:
    print("Pass")
else:
    print("Fail")
```

```python
number = 9

if number % 2 == 0:
    print("Even Number")
else:
    print("Odd Number")
```

```python
age = int(input("Enter your age: "))

if age >= 18:
    print("Eligible to Vote")
else:
    print("Not Eligible to Vote")
```

```python
number = int(input("Enter a number: "))

if number >= 0:
    print("Positive Number")
else:
    print("Negative Number")
```

```python
password = input("Enter Password: ")

if password == "python123":
    print("Login Successful")
else:
    print("Wrong Password")
```

```python
salary = int(input("Enter Salary: "))

if salary >= 50000:
    print("High Salary")
else:
    print("Average Salary")
```

## Flow of If...Else

Condition

↓

True?
    ↓
Execute if block

False?
    ↓
Execute else block

## Common Mistakes

❌ Forgetting the colon (:)

Wrong

if age >= 18

Correct

if age >= 18:

---

❌ Wrong indentation

Wrong

if age >= 18:
print("Eligible")

Correct

if age >= 18:
    print("Eligible")

```python
age = 16

if age >= 18:
    print("Adult")
else:
    print("Minor")

print("Program Ended")
```

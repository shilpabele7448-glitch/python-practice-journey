## What is Match-Case?

The match-case statement compares a value against different cases.

When a matching case is found, that block of code is executed.

It is similar to the switch statement available in many other programming languages.

## Syntax

match variable:

    case value1:
        statement

    case value2:
        statement

    case _:
        default statement

```python
day = int(input("Enter Day Number (1-7): "))

match day:

    case 1:
        print("Monday")

    case 2:
        print("Tuesday")

    case 3:
        print("Wednesday")

    case 4:
        print("Thursday")

    case 5:
        print("Friday")

    case 6:
        print("Saturday")

    case 7:
        print("Sunday")

    case _:
        print("Invalid Day")
```

```python
num1 = int(input("Enter First Number: "))
num2 = int(input("Enter Second Number: "))
operator = input("Enter Operator (+, -, *, /): ")

match operator:

    case "+":
        print("Result =", num1 + num2)

    case "-":
        print("Result =", num1 - num2)

    case "*":
        print("Result =", num1 * num2)

    case "/":
        print("Result =", num1 / num2)

    case _:
        print("Invalid Operator")
```

```python
signal = input("Enter Signal Color: ")

match signal:

    case "Red":
        print("Stop")

    case "Yellow":
        print("Wait")

    case "Green":
        print("Go")

    case _:
        print("Invalid Signal")
```

```python
month = int(input("Enter Month Number: "))

match month:

    case 1:
        print("January")

    case 2:
        print("February")

    case 3:
        print("March")

    case 4:
        print("April")

    case 5:
        print("May")

    case 6:
        print("June")

    case 7:
        print("July")

    case 8:
        print("August")

    case 9:
        print("September")

    case 10:
        print("October")

    case 11:
        print("November")

    case 12:
        print("December")

    case _:
        print("Invalid Month")
```

```python
choice = int(input("Enter Choice (1-3): "))

match choice:

    case 1:
        print("Add New Student")

    case 2:
        print("View Student Details")

    case 3:
        print("Exit")

    case _:
        print("Invalid Choice")
```

```python
grade = input("Enter Grade: ")

match grade:

    case "A":
        print("Excellent")

    case "B":
        print("Very Good")

    case "C":
        print("Good")

    case "D":
        print("Pass")

    case "F":
        print("Fail")

    case _:
        print("Invalid Grade")
```

## Advantages of Match-Case

- Code is cleaner.
- Easy to read.
- Better than long if...elif...else statements for many choices.
- Improves code organization.

## Common Mistakes

❌ Using match in Python versions below 3.10.

❌ Forgetting the default case.

Always use:

case _:

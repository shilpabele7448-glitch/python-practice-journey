## Why is this Topic Important?

Sometimes one condition depends on another.

For example:

- First check whether a student has passed.
- If passed, then check whether the student scored above 90.

This is where Nested If is useful.

## What is a Nested If?

A Nested If means writing one if statement inside another if statement.

The inner if runs only if the outer if condition is True.

## Syntax

if condition1:

    if condition2:
        statement

```python
marks = int(input("Enter Marks: "))

if marks >= 35:

    print("Pass")

    if marks >= 90:
        print("Excellent Performance")
```

```python
age = int(input("Enter Age: "))

if age >= 18:

    print("Eligible to Vote")

    if age >= 60:
        print("Senior Citizen")
```

```python
balance = 10000
withdraw = int(input("Enter Amount: "))

if withdraw <= balance:

    print("Sufficient Balance")

    if withdraw <= 5000:
        print("Transaction Successful")
```

```python
username = input("Username: ")
password = input("Password: ")

if username == "admin":

    if password == "python123":
        print("Login Successful")
```

```python
experience = int(input("Experience: "))
salary = int(input("Salary: "))

if experience >= 5:

    if salary >= 50000:
        print("Eligible for Bonus")
```

```python
number = int(input("Enter Number: "))

if number > 0:

    if number % 2 == 0:
        print("Positive Even Number")
```

## Flow of Nested If

Condition 1

↓

True?

↓

Condition 2

↓

True?

↓

Execute Code

## Common Mistakes

## Why is this Topic Important?

Sometimes there are more than two possible outcomes.

Examples:
- Student Grades (A, B, C, D, Fail)
- Traffic Signals
- Income Tax Slabs
- Employee Bonus

For these situations, Python provides the if...elif...else statement.

## What is if...elif...else?

The if statement checks the first condition.

If it is False, Python checks the elif condition.

Python continues checking each elif until one condition becomes True.

If none of the conditions are True, the else block executes.

```python
marks = int(input("Enter Marks: "))

if marks >= 90:
    print("Grade A")

elif marks >= 75:
    print("Grade B")

elif marks >= 50:
    print("Grade C")

elif marks >= 35:
    print("Grade D")

else:
    print("Fail")
```

```python
age = int(input("Enter Age: "))

if age < 13:
    print("Child")

elif age < 20:
    print("Teenager")

elif age < 60:
    print("Adult")

else:
    print("Senior Citizen")
```

```python
signal = input("Enter Signal Color: ")

if signal == "Red":
    print("Stop")

elif signal == "Yellow":
    print("Wait")

elif signal == "Green":
    print("Go")

else:
    print("Invalid Signal")
```

```python
salary = int(input("Enter Salary: "))

if salary >= 100000:
    print("High Income")

elif salary >= 50000:
    print("Medium Income")

elif salary >= 25000:
    print("Low Income")

else:
    print("Very Low Income")
```

```python
day = int(input("Enter Day Number (1-7): "))

if day == 1:
    print("Monday")

elif day == 2:
    print("Tuesday")

elif day == 3:
    print("Wednesday")

elif day == 4:
    print("Thursday")

elif day == 5:
    print("Friday")

elif day == 6:
    print("Saturday")

elif day == 7:
    print("Sunday")

else:
    print("Invalid Day")
```

```python
bmi = float(input("Enter BMI: "))

if bmi < 18.5:
    print("Underweight")

elif bmi < 25:
    print("Normal Weight")

elif bmi < 30:
    print("Overweight")

else:
    print("Obese")
```

## Flow of if...elif...else

Start

↓

Check first condition

↓

True?

Yes → Execute and Stop

↓

No

↓

Check next elif

↓

Continue until one condition is True

↓

If none are True → Execute else

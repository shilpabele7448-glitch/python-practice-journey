# Nested Loops

## Topic
Nested Loops

## Description

A Nested Loop is a loop inside another loop.

The outer loop executes first, and for each iteration of the outer loop, the inner loop executes completely.

**Author:** Shilpa

## Why Do We Use Nested Loops?

Nested loops are useful when we need to work with:

- Patterns
- Tables
- Matrices
- Grids
- Rows and Columns

## Syntax

```python
for outer in range(...):

    for inner in range(...):
        # Code

```

````python
for i in range(3):
    for j in range(3):
        print(i, j)
````

````python
for i in range(1, 4):
    print("Row", i)

    for j in range(1, 4):
        print("Column", j)
````

````python
for i in range(1, 6):

    for j in range(1, 6):
        print("*", end=" ")

    print()
````

````python
for i in range(1, 6):

    for j in range(1, i + 1):
        print("*", end=" ")

    print()
````

````python
for i in range(5, 0, -1):

    for j in range(i):
        print("*", end=" ")

    print()
````

````python
for i in range(1, 4):

    for j in range(1, 4):
        print(i * j, end="\t")

    print()
````

````python
for i in range(1, 6):

    for j in range(1, 11):
        print(i, "x", j, "=", i * j)

    print("-" * 20)
````

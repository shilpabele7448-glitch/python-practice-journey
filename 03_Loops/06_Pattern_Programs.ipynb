## What is Pattern Programming?

Pattern programming means printing stars (`*`), numbers, or characters in a specific shape using loops.

It mainly uses nested loops.

## Pattern 1 - Square Star Pattern

```python
for i in range(5):
    for j in range(5):
        print("*", end=" ")
    print()
```

## Pattern 2 - Right Triangle

```python
for i in range(1, 6):
    for j in range(i):
        print("*", end=" ")
    print()
```

## Pattern 3 - Inverted Triangle

```python
for i in range(5, 0, -1):
    for j in range(i):
        print("*", end=" ")
    print()
```

## Pattern 4 - Number Triangle

```python
for i in range(1, 6):
    for j in range(1, i + 1):
        print(j, end=" ")
    print()
```

## Pattern 5 - Repeated Number Triangle

```python
for i in range(1, 6):
    for j in range(i):
        print(i, end=" ")
    print()
```

## Pattern 6 - Floyd's Triangle

```python
num = 1

for i in range(1, 6):
    for j in range(i):
        print(num, end=" ")
        num += 1
    print()
```

## Pattern 7 - Alphabet Triangle

```python
for i in range(65, 70):
    for j in range(65, i + 1):
        print(chr(j), end=" ")
    print()
```

## Pattern 8 - Multiplication Table Grid

```python
for i in range(1, 6):
    for j in range(1, 6):
        print(i * j, end="\t")
    print()
```

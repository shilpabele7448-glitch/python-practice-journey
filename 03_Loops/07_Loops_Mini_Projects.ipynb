# Mini Project 1 - Multiplication Table Generator

```python
number = int(input("Enter a number: "))

for i in range(1, 11):
    print(f"{number} x {i} = {number * i}")
```

# Mini Project 2 - Sum of Numbers

```python
n = int(input("Enter a number: "))

total = 0

for i in range(1, n + 1):
    total += i

print("Sum =", total)
```

# Mini Project 3 - Factorial Calculator

```python
number = int(input("Enter a number: "))

factorial = 1

for i in range(1, number + 1):
    factorial *= i

print("Factorial =", factorial)
```

# Mini Project 4 - Count Digits

```python
number = input("Enter a number: ")

count = 0

for digit in number:
    count += 1

print("Total Digits =", count)
```

# Mini Project 5 - Password Checker

```python
correct_password = "python123"

while True:
    password = input("Enter Password: ")

    if password == correct_password:
        print("Login Successful")
        break

    print("Wrong Password")
```

# Mini Project 6 - Number Guessing Game

```python
secret = 7

while True:
    guess = int(input("Guess the number: "))

    if guess == secret:
        print("Correct Guess!")
        break

    print("Try Again!")
```

# Mini Project 7 - Print Even Numbers

```python
for i in range(1, 21):

    if i % 2 != 0:
        continue

    print(i)
```

# Mini Project 8 - Reverse Counting

```python
for i in range(10, 0, -1):
    print(i)

print("Blast Off!")
```

# Mini Project 9 - Average of Numbers

```python
total = 0

for i in range(1, 6):
    number = int(input(f"Enter number {i}: "))
    total += number

average = total / 5

print("Average =", average)
```

# Mini Project 10 - Star Pattern

```python
rows = 5

for i in range(1, rows + 1):
    for j in range(i):
        print("*", end=" ")
    print()
```

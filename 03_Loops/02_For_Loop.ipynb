## Description

This notebook explains how to use the **for loop** in Python.

A for loop is used to repeat a block of code for every item in a sequence like a string, list, tuple, or range.

**Author:** Shilpa

## What is a For Loop?

A **for loop** is used when we know how many times we want to repeat something.

### Syntax

```python
for variable in sequence:
    # code
```

````python
for i in range(5):
    print(i)
````

````python
for i in range(1, 6):
    print(i)
````

````python
for i in range(2, 11, 2):
    print(i)
````

````python
name = "Shilpa"

for letter in name:
    print(letter)
````

````python
fruits = ["Apple", "Banana", "Mango"]

for fruit in fruits:
    print(fruit)
````

````python
for i in range(1, 6):
    print("Hello")
````

````python
for i in range(1, 11):
    print(i, end=" ")
````

## Important Points

- A for loop repeats code automatically.
- `range()` is commonly used with for loops.
- A for loop can iterate through strings, lists, tuples, sets, and dictionaries.
- Indentation is required inside the loop.

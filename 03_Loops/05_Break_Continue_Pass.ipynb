## Why Do We Need Loop Control Statements?

Sometimes we don't want a loop to execute every iteration in the same way.

We may want to:

- Stop the loop completely.
- Skip one iteration.
- Keep a placeholder for future code.

Python provides:

- break
- continue
- pass

## 1. Break Statement

The `break` statement immediately terminates the loop.

Once `break` is executed, the loop stops and the program continues with the next statement after the loop.

```python
for i in range(1, 11):
    if i == 6:
        break
    print(i)
```

### Real-Life Example of Break

Imagine searching for your friend's name in a class list.

Once you find the name, you stop searching.

This is similar to the `break` statement.

```python
password = "python123"

while True:
    user = input("Enter Password: ")

    if user == password:
        print("Login Successful")
        break

    print("Wrong Password")
```

## 2. Continue Statement

The `continue` statement skips the current iteration and moves to the next iteration of the loop.

```python
for i in range(1, 11):

    if i == 5:
        continue

    print(i)
```

```python
for i in range(1, 11):

    if i % 2 == 0:
        continue

    print(i)
```

### Real-Life Example of Continue

Imagine a teacher checking attendance.

If one student is absent, the teacher skips that student and continues checking the remaining students.

## 3. Pass Statement

The `pass` statement does nothing.

It is used as a placeholder when code will be added later.

```python
for i in range(5):

    if i == 3:
        pass

    print(i)
```

```python
name = "Shilpa"

if len(name) > 5:
    pass
```

## Difference Between break, continue and pass

| Statement | Purpose |
|-----------|---------|
| break | Stops the loop completely |
| continue | Skips the current iteration |
| pass | Does nothing (placeholder) |

## Data Analytics Connection

These statements are useful while processing datasets.

Examples:

- `break` → Stop processing when invalid data is found.
- `continue` → Skip missing or incorrect records.
- `pass` → Leave unfinished logic while developing a program.

```python
marks = [75, 82, -1, 90, 65]

for mark in marks:

    if mark == -1:
        continue

    print(mark)
```

## Common Mistakes

❌ Using `break` when you only want to skip one iteration.

❌ Forgetting that `continue` skips the remaining code in the current iteration.

❌ Thinking `pass` stops execution (it does not).

## Key Takeaways

- `break` stops the loop.
- `continue` skips one iteration.
- `pass` is a placeholder.
- These statements improve control over loops.

# Conclusion

The `break`, `continue`, and `pass` statements make loops more flexible and efficient. Understanding when to use each one helps you write cleaner and more effective Python programs.

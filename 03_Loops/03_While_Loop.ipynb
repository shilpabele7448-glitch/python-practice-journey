# While Loop

## Topic
While Loop

## Description

A while loop executes a block of code repeatedly as long as the given condition is True.

Unlike the for loop, the number of iterations is usually not known in advance.

**Author:** Shilpa

## Syntax

```python
count = 1

while count <= 5:
    print(count)
    count += 1
```

## Example 1 - Print Hello World 5 Times

```python
count = 1

while count <= 5:
    print("Hello World")
    count += 1
```

## Example 2 - Print Numbers from 10 to 1

```python
num = 10

while num >= 1:
    print(num)
    num -= 1
```

## Example 3 - Sum of First 10 Numbers

```python
count = 1
total = 0

while count <= 10:
    total += count
    count += 1

print("Sum =", total)
```

## Example 4 - Multiplication Table

```python
num = 7
i = 1

while i <= 10:
    print(num, "x", i, "=", num * i)
    i += 1
```

## Example 5 - Password Check

```python
password = ""

while password != "python123":
    password = input("Enter Password: ")

print("Access Granted")
```

```python
#Reverse a number.
i=10
while(i<=1):
    print(i)
    i=i+1
print("End")
```

```python
#Reverse a number.
num = 12345
reversed_num = 0

while num > 0:
    digit = num % 10                # Extracts the last digit (e.g., 5)
    reversed_num = (reversed_num * 10) + digit  # Appends digit to the new number
    num = num // 10  
    print(reversed_num)
```

```python
#Sum of digits
num = 12345
digit_sum = 0

while num > 0:
    digit_sum += num % 10  # Extracts the last digit
    num //= 10             # Removes the last digit

print(digit_sum)  
```

## Common Mistake

Forgetting to update the loop variable can create an infinite loop.

## Key Points

- while loop runs until the condition becomes False.
- It is useful when the number of iterations is not fixed.
- Always update the loop variable.
- Be careful to avoid infinite loops.

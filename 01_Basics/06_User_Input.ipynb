## What is User Input?

User input means entering data while the program is running.

Python uses the `input()` function to receive data from the user.

## Syntax

variable_name = input("Enter something: ")

````python
name = input("Enter your name: ")

print("Hello,", name)
````

````python
city = input("Enter your city: ")

print("City:", city)
````

````python
age = input("Enter your age: ")

print(age)
print(type(age))
````

````python
age = int(input("Enter your age: "))

print(age)
print(type(age))
````

````python
height = float(input("Enter your height: "))

print(height)
print(type(height))
````

````python
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

print("Sum =", num1 + num2)
````

````python
name = input("Enter your name: ")
age = int(input("Enter your age: "))

print("Name:", name)
print("Age:", age)
````

## Why Do We Convert Input?

The input() function always returns a string.

If we want to perform calculations, we must convert the input into int or float.

````python
a = input("Enter first number: ")
b = input("Enter second number: ")

print(a + b)
````

````python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

print(a + b)
````

## Common Mistakes

❌ Forgetting to convert input before calculations.

Wrong:

```python
age = input("Enter age: ")
print(age + 10)
```

Correct:

```python
age = int(input("Enter age: "))
print(age + 10)
```

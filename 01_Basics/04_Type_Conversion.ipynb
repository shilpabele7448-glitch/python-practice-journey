## What is Type Conversion?

Type Conversion means changing one data type into another.

For example:
- Convert an integer into a float.
- Convert a float into an integer.
- Convert a number into a string.

## Why Do We Need Type Conversion?

Sometimes, the data we receive is not in the format we need.

For example:
- User input is received as text.
- We may need to convert it into a number before performing calculations.

Type conversion helps us perform the required operations correctly.

```python
age = 22

print(age)
print(type(age))
```

```python
age = 22

new_age = float(age)

print(new_age)
print(type(new_age))
```

```python
price = 99.99

new_price = int(price)

print(new_price)
print(type(new_price))
```

```python
number = 100

text = str(number)

print(text)
print(type(text))
```

```python
print(bool(1))
print(bool(0))
```

```python
a = "25"
b = "10"

print(a + b)
```

```python
a = int("25")
b = int("10")

print(a + b)
```

## Common Type Conversion Functions

- int() → Converts to Integer
- float() → Converts to Float
- str() → Converts to String
- bool() → Converts to Boolean

```python
print(int(25.8))
print(float(25))
print(str(25))
print(bool(25))
```
